```python
# ------------ Tutorial 03 ----------------
# Part 1 - Google Colab and Python

import numpy as np

# 1. Create a numpy array of size 2 x 2
array_2x2 = np.array([[1, 2], [3, 4]])
print("Array 2x2:\n", array_2x2)

# 2. Create a random numpy array of size 3 x 4
random_array_3x4 = np.random.rand(3, 4)
print("\nRandom array 3x4:\n", random_array_3x4)

# 3. Create a numpy array of ones and a numpy array of zeroes
ones_array = np.ones((3, 3))
zeros_array = np.zeros((3, 3))
print("\nArray of ones:\n", ones_array)
print("\nArray of zeroes:\n", zeros_array)

# 4. Create a random numpy array of size 6x3 and transpose it
random_array_6x3 = np.random.rand(6, 3)
print("\nRandom array 6x3:\n", random_array_6x3)
print("\nTransposed array 3x6:\n", random_array_6x3.T)

# 5. Create two random numpy arrays of size n x p and p x m and compute the dot product
n, p, m = 2, 3, 4
A = np.random.rand(n, p)
B = np.random.rand(p, m)
dot_product = np.dot(A, B)
print("\nMatrix A (n x p):\n", A)
print("\nMatrix B (p x m):\n", B)
print("\nDot product (n x m):\n", dot_product)
```
